# 🔄 Словарь двойственности

Парад двойственных категорных конструкций. Одна ось — обращение стрелок `C &#8644; C^op`.
Подача: разминка на **примитивах** &#8594; рано вводится **движок — расширения Кана** &#8594;
и дальше пределы, сопряжения, монады/комонады выводятся **через Кан**.


## 0. Мотивация: что такое двойственность

**Двойственность = обращение всех стрелок.** Каждой конструкции в категории `C` отвечает
её зеркальное отражение в противоположной категории `C^op` (те же объекты, стрелки развёрнуты).
Приставка «co-» буквально значит «то же самое в `C^op`»: ко-предел, ко-произведение, ко-монада.

Дальше мы увидим, что бо&#769;льшая часть конструкций — **частные случаи расширений Кана**
(`Lan`/`Ran`), которые сами зеркальны друг другу. Но сперва — чистые примитивы, которые
Каном не порождаются: моно/эпи, вариантность функтора, алгебры/коалгебры.

## Мономорфизм &#8611; / Эпиморфизм &#8608; (примитив)

**Идея.** Чистая `C^op`-двойственность: мономорфизм в `C` — это в точности эпиморфизм в `C^op` (и наоборот). Базовые понятия на языке стрелок; расширениями Кана **не** порождаются.

**Определение (через сократимость, в любой категории).** Пусть `f : a → b`.
- `f` — **мономорфизм** (`a &#8611; b`), если он *лево-сократим*: из `f∘g = f∘h` следует `g = h`.
- `f` — **эпиморфизм** (`a &#8608; b`), если он *право-сократим*: из `g∘f = h∘f` следует `g = h`.

Одно переходит в другое **обращением всех стрелок** — это и есть двойственность.

**Тонкость — почему НЕ «инъекция/сюръекция».** Инъективность/сюръективность — *точечные* понятия (нужны глобальные элементы `1 → X`), осмысленны лишь в конкретных / вполне-точечных (well-pointed) категориях:
- `mono ⟹ инъективна на точках` — **всегда** (представимый `Hom(1,−)` сохраняет мономорфизмы); обратное (`инъективна на точках ⟹ mono`) — только если `1` является генератором (как в **Set**).
- `epi ⟺ сюръекция` — связь ещё слабее: `Hom(1,−)` **не** сохраняет эпиморфизмы. Совпадают в **Set** (с аксиомой выбора), в общем случае — нет.

Поэтому определение даём через сократимость, а инъекцию/сюръекцию — лишь как «тень» в Set.

**В чём двойственность:** `mono` и `epi` — зеркало друг друга относительно `C ⇄ C^op` (обращение стрелок).

**Где в курсе:** [Toposes](Toposes.ipynb) — классификатор подобъектов `s &#8611; Ω`: мономорфизмы = подобъекты.

```haskell
-- В Hask (Set-подобная категория) mono ~ инъекция, epi ~ сюръекция.
-- Категорное определение mono/epi -- через сократимость (см. markdown выше);
-- здесь -- его "тень" на конечных типах.
import Data.List (nub)

-- инъективность: разные входы дают разные выходы
injective :: Eq b => [a] -> (a -> b) -> Bool
injective dom f = let ys = map f dom in length ys == length (nub ys)

-- сюръективность: весь кодомен покрыт
surjective :: Eq b => [a] -> [b] -> (a -> b) -> Bool
surjective dom cod f = all (\y -> y `elem` map f dom) cod

toInt :: Bool -> Int
toInt b = if b then 1 else 0

isPos :: Int -> Bool
isPos n = n > 0

-- (mono ~ injective, epi ~ surjective)
(injective [False, True] toInt, surjective [-2, -1, 0, 1, 2] [False, True] isPos)
```

## Ковариантный / Контравариантный функтор (примитив)

**Идея.** Функтор переносит стрелки между категориями. **Ковариантный** сохраняет направление (`fmap`), **контравариантный** — обращает (`contramap`). И это снова чистая `C^op`-двойственность: контравариантный функтор `F : C → D` — это в точности ковариантный функтор `F : C^op → D`.

**Определение.** На морфизме `f : a → b`:
- ковариантный: `F f : F a → F b` — `fmap :: (a → b) → (F a → F b)`;
- контравариантный: `F f : F b → F a` — `contramap :: (a → b) → (F b → F a)`.

Нижняя строка `D` показывает обращение: слева `Fa → Fb` (то же направление, что `f`), справа `Fb → Fa` (обращённое). Это и есть переход к `C^op`.

**В чём двойственность:** контравариантный функтор = ковариантный функтор на противоположной категории (`F : C^op → D`).

**Где в курсе:** [FunctorHierarchy](FunctorHierarchy.ipynb) — Functor / Contravariant / Bifunctor.

```haskell
-- Ковариантный функтор: fmap сохраняет направление стрелки (a->b  даёт  Fa->Fb).
exCov :: Maybe Int
exCov = fmap (+1) (Just 10)   -- Just 11

-- Контравариантный функтор: contramap обращает направление (a->b  даёт  Fb->Fa).
-- (Data.Functor.Contravariant может отсутствовать в старом GHC -- объявим класс локально.)
class Contravariant f where
  contramap :: (a -> b) -> f b -> f a

newtype Predicate a = Predicate { runPred :: a -> Bool }

instance Contravariant Predicate where
  contramap g (Predicate p) = Predicate (p . g)

isEven :: Predicate Int
isEven = Predicate even

-- предикат на Int тянется на String через  length :: String -> Int  (обращение стрелки)
lenEven :: Predicate String
lenEven = contramap length isEven

-- (Just 11, "hello"->длина 5 нечётна, "hi"->длина 2 чётна)
(exCov, runPred lenEven "hello", runPred lenEven "hi")
```

## Алгебра / Коалгебра (примитив)

**Идея.** Зафиксируем эндофунктор `F`. **F-алгебра** — это `(a, f : F a → a)` (структура *входит* в носитель), **F-коалгебра** — `(a, g : a → F a)` (структура *выходит*). Одно переходит в другое обращением стрелки структуры. Это снова примитив: неподвижные точки эндофунктора, расширениями Кана **не** порождаются.

**Универсальное свойство — вот где вся суть.**
- **Начальная** F-алгебра `(μF, in)`: для *любой* алгебры `(a, f)` существует **единственный** гомоморфизм `h = cata f : μF → a`, замыкающий квадрат — `cata f ∘ in = f ∘ F(cata f)`.
- **Финальная** F-коалгебра `(νF, out)`: дуально — для любой коалгебры `(a, g)` единственный `h = ana g : a → νF`, `out ∘ ana g = F(ana g) ∘ g`.
- **Лемма Ламбека:** `in` и `out` — изоморфизмы, т.е. `μF ≅ F(μF)` (носитель — настоящая неподвижная точка `F`). `cata` = свёртка (fold), `ana` = развёртка (unfold).

**В чём двойственность:** финальная коалгебра — это начальная алгебра в `C^op` (обращаем каждую стрелку: `in` ↦ `out`, `cata` ↦ `ana`).

**Где в курсе:** [AlgebrasCoalgebras](AlgebrasCoalgebras.ipynb) — F-алгебры, катаморфизмы, схемы рекурсии.

```haskell
-- F-алгебра:  f a -> a ;  F-коалгебра:  a -> f a  (обращение стрелки структуры).
type Algebra   f a = f a -> a
type Coalgebra f a = a -> f a

-- неподвижная точка эндофунктора: носитель начальной алгебры / финальной коалгебры
newtype Fix f = In { out :: f (Fix f) }

-- единственный гомоморфизм ИЗ начальной алгебры (catamorphism / fold)
cata :: Functor f => Algebra f a -> Fix f -> a
cata alg = alg . fmap (cata alg) . out

-- единственный гомоморфизм В финальную коалгебру (anamorphism / unfold)
ana :: Functor f => Coalgebra f a -> a -> Fix f
ana coalg = In . fmap (ana coalg) . coalg

-- пример эндофунктора: натуральные числа
data NatF r = ZeroF | SuccF r deriving (Functor, Show)

toInt :: Algebra NatF Int        -- F a -> a
toInt ZeroF     = 0
toInt (SuccF n) = n + 1

fromInt :: Coalgebra NatF Int     -- a -> F a
fromInt 0 = ZeroF
fromInt n = SuccF (n - 1)

-- ana строит Nat из 5, cata сворачивает обратно -> 5
cata toInt (ana fromInt 5)
```

## Расширения Кана: Lan / Ran (движок)

**Идея.** Дан функтор `F : C → D` и функтор `K : C → C'`. Расширения Кана продолжают `F` вдоль `K` до функтора `C' → D` — наилучшим образом *слева* (`Lan`) и *справа* (`Ran`). Это зеркальная пара относительно `C^op` и одновременно **движок** всего словаря.

**Универсальное свойство (сопряжения с пред-композицией `−∘K`):**

`Lan_K ⊣ (−∘K) ⊣ Ran_K`, то есть естественные изоморфизмы
- `Nat(Lan_K F, G) ≅ Nat(F, G∘K)` — единица `η : F ⇒ (Lan_K F)∘K`;
- `Nat(G, Ran_K F) ≅ Nat(G∘K, F)` — коединица `ε : (Ran_K F)∘K ⇒ F`.

**«Всё — расширение Кана».** Дальше выводим через этот движок: пределы/копределы — Кан вдоль `J → 1` (§ ниже), сопряжения — `Ran_F Id` / `Lan_G Id`, монада коплотности — `Ran f f`, комонада плотности — `Lan f f`.

**В чём двойственность:** `Lan` (левое, копределоподобное) ↔ `Ran` (правое, пределоподобное) — зеркало относительно `C^op`.

**Где в курсе:** [KanExtensions](KanExtensions.ipynb) — Lan/Ran, Codensity, плотность, «все концепции — расширения Кана».

```haskell
-- Левое (Lan) и правое (Ran) расширения Кана как типы Haskell:
--   Lan: существует b  (экзистенциал);  Ran: для всех b  (forall)
data    Lan g h a = forall b. Lan (g b -> a) (h b)
newtype Ran g h a = Ran { runRan :: forall b. (a -> g b) -> h b }

instance Functor (Lan g h) where
  fmap f (Lan k hb) = Lan (f . k) hb

instance Functor (Ran g h) where
  fmap f (Ran r) = Ran (\k -> r (k . f))

-- Частный случай Ran f f -- монада коплотности (Codensity); превью раздела monad/comonad.
newtype Codensity m a = Codensity (forall b. (a -> m b) -> m b)

toCod :: Monad m => m a -> Codensity m a
toCod ma = Codensity (ma >>=)

fromCod :: Monad m => Codensity m a -> m a
fromCod (Codensity r) = r return

-- round-trip через Codensity (= Ran Maybe Maybe): Just 7
fromCod (toCod (Just 7)) :: Maybe Int
```

**В чём двойственность:** начальный объект `0` — это терминальный объект `1` в `C^op` (обращаем единственные стрелки).

**Где в курсе:** [TypeAlgebra](TypeAlgebra.ipynb) — `Void` (= `0`) и `()` (= `1`), кардинальности типов.

```haskell
import Data.Void (Void, absurd)

-- Начальный объект 0 = Void: ЕДИНСТВЕННАЯ стрелка в любой тип a.
fromInitial :: Void -> a
fromInitial = absurd

-- Терминальный объект 1 = (): ЕДИНСТВЕННАЯ стрелка из любого типа a.
toTerminal :: a -> ()
toTerminal _ = ()

-- демо единственности: куда ни шли -- результат один и тот же ()
(toTerminal (42 :: Int), toTerminal "hi")
```
