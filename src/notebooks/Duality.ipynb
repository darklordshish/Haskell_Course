# 🔄 Словарь двойственности

Парад двойственных категорных конструкций. Одна ось — обращение стрелок `C &#8644; C^op`.
Подача: разминка на **примитивах** &#8594; рано вводится **движок — расширения Кана** &#8594;
и дальше пределы, сопряжения, монады/комонады выводятся **через Кан**.


## 0. Мотивация: что такое двойственность

**Двойственность = обращение всех стрелок.** Каждой конструкции в категории `C` отвечает
её зеркальное отражение в противоположной категории `C^op` (те же объекты, стрелки развёрнуты).
Приставка «co-» буквально значит «то же самое в `C^op`»: ко-предел, ко-произведение, ко-монада.

Дальше мы увидим, что бо&#769;льшая часть конструкций — **частные случаи расширений Кана**
(`Lan`/`Ran`), которые сами зеркальны друг другу. Но сперва — чистые примитивы, которые
Каном не порождаются: моно/эпи, вариантность функтора, алгебры/коалгебры.

## Мономорфизм &#8611; / Эпиморфизм &#8608; (примитив)

**Идея.** Чистая `C^op`-двойственность: мономорфизм в `C` — это в точности эпиморфизм в `C^op` (и наоборот). Базовые понятия на языке стрелок; расширениями Кана **не** порождаются.

**Определение (через сократимость, в любой категории).** Пусть `f : A → B`.
- `f` — **мономорфизм** (`A &#8611; B`), если он *лево-сократим*: из `f∘g = f∘h` следует `g = h`.
- `f` — **эпиморфизм** (`A &#8608; B`), если он *право-сократим*: из `g∘f = h∘f` следует `g = h`.

Одно переходит в другое **обращением всех стрелок** — это и есть двойственность.

**Тонкость — почему НЕ «инъекция/сюръекция».** Инъективность/сюръективность — *точечные* понятия (нужны глобальные элементы `1 → X`), осмысленны лишь в конкретных / вполне-точечных (well-pointed) категориях:
- `mono ⟹ инъективна на точках` — **всегда** (представимый `Hom(1,−)` сохраняет мономорфизмы); обратное (`инъективна на точках ⟹ mono`) — только если `1` является генератором (как в **Set**).
- `epi ⟺ сюръекция` — связь ещё слабее: `Hom(1,−)` **не** сохраняет эпиморфизмы. Совпадают в **Set** (с аксиомой выбора), в общем случае — нет.

Поэтому определение даём через сократимость, а инъекцию/сюръекцию — лишь как «тень» в Set.

**В чём двойственность:** `mono` и `epi` — зеркало друг друга относительно `C ⇄ C^op` (обращение стрелок).

**Где в курсе:** [Toposes](Toposes.ipynb) — классификатор подобъектов `S &#8611; Ω`: мономорфизмы = подобъекты.

```haskell
-- В Hask (Set-подобная категория) mono ~ инъекция, epi ~ сюръекция.
-- Категорное определение mono/epi -- через сократимость (см. markdown выше);
-- здесь -- его "тень" на конечных типах.
import Data.List (nub)

-- инъективность: разные входы дают разные выходы
injective :: Eq b => [a] -> (a -> b) -> Bool
injective dom f = let ys = map f dom in length ys == length (nub ys)

-- сюръективность: весь кодомен покрыт
surjective :: Eq b => [a] -> [b] -> (a -> b) -> Bool
surjective dom cod f = all (\y -> y `elem` map f dom) cod

toInt :: Bool -> Int
toInt b = if b then 1 else 0

isPos :: Int -> Bool
isPos n = n > 0

-- (mono ~ injective, epi ~ surjective)
(injective [False, True] toInt, surjective [-2, -1, 0, 1, 2] [False, True] isPos)
```
