# 🔄 Словарь двойственности

Парад двойственных категорных конструкций. Одна ось — обращение стрелок `C &#8644; C^op`.
Подача: разминка на **примитивах** &#8594; рано вводится **движок — расширения Кана** &#8594;
и дальше пределы, сопряжения, монады/комонады выводятся **через Кан**.


## 0. Мотивация: что такое двойственность

**Двойственность = обращение всех стрелок.** Каждой конструкции в категории `C` отвечает
её зеркальное отражение в противоположной категории `C^op` (те же объекты, стрелки развёрнуты).
Приставка «co-» буквально значит «то же самое в `C^op`»: ко-предел, ко-произведение, ко-монада.

Дальше мы увидим, что бо&#769;льшая часть конструкций — **частные случаи расширений Кана**
(`Lan`/`Ran`), которые сами зеркальны друг другу. Но сперва — чистые примитивы, которые
Каном не порождаются: моно/эпи, вариантность функтора, алгебры/коалгебры.

<a id="dual-monoepi"></a>
## Мономорфизм &#8611; / Эпиморфизм &#8608; (примитив)

**Идея.** Чистая `C^op`-двойственность: мономорфизм в `C` — это в точности эпиморфизм в `C^op` (и наоборот). Базовые понятия на языке стрелок; расширениями Кана **не** порождаются.

**Определение (через сократимость, в любой категории).** Пусть `f : a → b`.
- `f` — **мономорфизм** (`a &#8611; b`), если он *лево-сократим*: из `f∘g = f∘h` следует `g = h`.
- `f` — **эпиморфизм** (`a &#8608; b`), если он *право-сократим*: из `g∘f = h∘f` следует `g = h`.

Одно переходит в другое **обращением всех стрелок** — это и есть двойственность.

![monoepi fig.1](../diagrams/duality/monoepi_1.svg)

**Тонкость — почему НЕ «инъекция/сюръекция».** Инъективность/сюръективность — *точечные* понятия (нужны глобальные элементы `1 → X`), осмысленны лишь в конкретных / вполне-точечных (well-pointed) категориях:
- `mono ⟹ инъективна на точках` — **всегда** (представимый `Hom(1,−)` сохраняет мономорфизмы); обратное (`инъективна на точках ⟹ mono`) — только если `1` является генератором (как в **Set**).
- `epi ⟺ сюръекция` — связь ещё слабее: `Hom(1,−)` **не** сохраняет эпиморфизмы. Совпадают в **Set** (с аксиомой выбора), в общем случае — нет.

Поэтому определение даём через сократимость, а инъекцию/сюръекцию — лишь как «тень» в Set.

**В чём двойственность:** `mono` и `epi` — зеркало друг друга относительно `C ⇄ C^op` (обращение стрелок).

**Где в курсе:** [Toposes](Toposes.ipynb) — классификатор подобъектов `s &#8611; Ω`: мономорфизмы = подобъекты.

```haskell
-- В Hask (Set-подобная категория) mono ~ инъекция, epi ~ сюръекция.
-- Категорное определение mono/epi -- через сократимость (см. markdown выше);
-- здесь -- его "тень" на конечных типах.
import Data.List (nub)

-- инъективность: разные входы дают разные выходы
injective :: Eq b => [a] -> (a -> b) -> Bool
injective dom f = let ys = map f dom in length ys == length (nub ys)

-- сюръективность: весь кодомен покрыт
surjective :: Eq b => [a] -> [b] -> (a -> b) -> Bool
surjective dom cod f = all (\y -> y `elem` map f dom) cod

boolToInt :: Bool -> Int
boolToInt b = if b then 1 else 0

isPos :: Int -> Bool
isPos n = n > 0

-- (mono ~ injective, epi ~ surjective)
(injective [False, True] boolToInt, surjective [-2, -1, 0, 1, 2] [False, True] isPos)
```

<a id="dual-variance"></a>
## Ковариантный / Контравариантный функтор (примитив)

**Идея.** Функтор переносит стрелки между категориями. **Ковариантный** сохраняет направление (`fmap`), **контравариантный** — обращает (`contramap`). И это снова чистая `C^op`-двойственность: контравариантный функтор `F : C → D` — это в точности ковариантный функтор `F : C^op → D`.

**Определение.** На морфизме `f : a → b`:
- ковариантный: `F f : F a → F b` — `fmap :: (a → b) → (F a → F b)`;
- контравариантный: `F f : F b → F a` — `contramap :: (a → b) → (F b → F a)`.

![variance fig.1](../diagrams/duality/variance_1.svg)

Нижняя строка `D` показывает обращение: слева `Fa → Fb` (то же направление, что `f`), справа `Fb → Fa` (обращённое). Это и есть переход к `C^op`.

**В чём двойственность:** контравариантный функтор = ковариантный функтор на противоположной категории (`F : C^op → D`).

**Где в курсе:** [FunctorHierarchy](FunctorHierarchy.ipynb) — Functor / Contravariant / Bifunctor.

```haskell
-- Ковариантный функтор: fmap сохраняет направление стрелки (a->b  даёт  Fa->Fb).
exCov :: Maybe Int
exCov = fmap (+1) (Just 10)   -- Just 11

-- Контравариантный функтор: contramap обращает направление (a->b  даёт  Fb->Fa).
-- (Data.Functor.Contravariant может отсутствовать в старом GHC -- объявим класс локально.)
class Contravariant f where
  contramap :: (a -> b) -> f b -> f a

newtype Predicate a = Predicate { runPred :: a -> Bool }

instance Contravariant Predicate where
  contramap g (Predicate p) = Predicate (p . g)

isEven :: Predicate Int
isEven = Predicate even

-- предикат на Int тянется на String через  length :: String -> Int  (обращение стрелки)
lenEven :: Predicate String
lenEven = contramap length isEven

-- (Just 11, "hello"->длина 5 нечётна, "hi"->длина 2 чётна)
(exCov, runPred lenEven "hello", runPred lenEven "hi")
```

<a id="dual-algcoalg"></a>
## Алгебра / Коалгебра (примитив)

**Идея.** Зафиксируем эндофунктор `F`. **F-алгебра** — это `(a, f : F a → a)` (структура *входит* в носитель), **F-коалгебра** — `(a, g : a → F a)` (структура *выходит*). Одно переходит в другое обращением стрелки структуры. Это снова примитив: неподвижные точки эндофунктора, расширениями Кана **не** порождаются.

**Универсальное свойство — вот где вся суть.**
- **Начальная** F-алгебра `(μF, in)`: для *любой* алгебры `(a, f)` существует **единственный** гомоморфизм `h = cata f : μF → a`, замыкающий квадрат — `cata f ∘ in = f ∘ F(cata f)`.
- **Финальная** F-коалгебра `(νF, out)`: дуально — для любой коалгебры `(a, g)` единственный `h = ana g : a → νF`, `out ∘ ana g = F(ana g) ∘ g`.
- **Лемма Ламбека:** `in` и `out` — изоморфизмы, т.е. `μF ≅ F(μF)` (носитель — настоящая неподвижная точка `F`). `cata` = свёртка (fold), `ana` = развёртка (unfold).

![algcoalg fig.1](../diagrams/duality/algcoalg_1.svg)

**В чём двойственность:** финальная коалгебра — это начальная алгебра в `C^op` (обращаем каждую стрелку: `in` ↦ `out`, `cata` ↦ `ana`).

**Где в курсе:** [AlgebrasCoalgebras](AlgebrasCoalgebras.ipynb) — F-алгебры, катаморфизмы, схемы рекурсии.

```haskell
-- F-алгебра:  f a -> a ;  F-коалгебра:  a -> f a  (обращение стрелки структуры).
type Algebra   f a = f a -> a
type Coalgebra f a = a -> f a

-- неподвижная точка эндофунктора: носитель начальной алгебры / финальной коалгебры
newtype Fix f = In { out :: f (Fix f) }

-- единственный гомоморфизм ИЗ начальной алгебры (catamorphism / fold)
cata :: Functor f => Algebra f a -> Fix f -> a
cata alg = alg . fmap (cata alg) . out

-- единственный гомоморфизм В финальную коалгебру (anamorphism / unfold)
ana :: Functor f => Coalgebra f a -> a -> Fix f
ana coalg = In . fmap (ana coalg) . coalg

-- пример эндофунктора: натуральные числа
data NatF r = ZeroF | SuccF r deriving (Functor, Show)

toInt :: Algebra NatF Int        -- F a -> a
toInt ZeroF     = 0
toInt (SuccF n) = n + 1

fromInt :: Coalgebra NatF Int     -- a -> F a
fromInt 0 = ZeroF
fromInt n = SuccF (n - 1)

-- ana строит Nat из 5, cata сворачивает обратно -> 5
cata toInt (ana fromInt 5)
```

<a id="dual-kan"></a>
## Расширения Кана: Lan / Ran (движок)

**Идея.** Дан функтор `F : C → D` и функтор `K : C → C'`. Расширения Кана продолжают `F` вдоль `K` до функтора `C' → D` — наилучшим образом *слева* (`Lan`) и *справа* (`Ran`). Это зеркальная пара относительно `C^op` и одновременно **движок** всего словаря.

**Универсальное свойство (сопряжения с пред-композицией `−∘K`):**

`Lan_K ⊣ (−∘K) ⊣ Ran_K`, то есть естественные изоморфизмы
- `Nat(Lan_K F, G) ≅ Nat(F, G∘K)` — единица `η : F ⇒ (Lan_K F)∘K`;
- `Nat(G, Ran_K F) ≅ Nat(G∘K, F)` — коединица `ε : (Ran_K F)∘K ⇒ F`.

![kan fig.1](../diagrams/duality/kan_1.svg)

**Зеркало `C^op`:** `Lan` и `Ran` — отражение друг друга: `(Lan_K F)^op ≅ Ran_(K^op)(F^op)`. Одна конструкция, два направления.

**«Всё — расширение Кана».** Дальше выводим через этот движок: пределы/копределы — Кан вдоль `J → 1` (§ ниже), сопряжения — `Ran_F Id` / `Lan_G Id`, монада коплотности — `Ran f f`, комонада плотности — `Lan f f`.

**В чём двойственность:** `Lan` (левое, копределоподобное) ↔ `Ran` (правое, пределоподобное) — зеркало относительно `C^op`.

**Где в курсе:** [KanExtensions](KanExtensions.ipynb) — Lan/Ran, Codensity, плотность, «все концепции — расширения Кана».

```haskell
-- Левое (Lan) и правое (Ran) расширения Кана как типы Haskell:
--   Lan: существует b  (экзистенциал);  Ran: для всех b  (forall)
data    Lan g h a = forall b. Lan (g b -> a) (h b)
newtype Ran g h a = Ran { runRan :: forall b. (a -> g b) -> h b }

instance Functor (Lan g h) where
  fmap f (Lan k hb) = Lan (f . k) hb

instance Functor (Ran g h) where
  fmap f (Ran r) = Ran (\k -> r (k . f))

-- Частный случай Ran f f -- монада коплотности (Codensity); превью раздела monad/comonad.
newtype Codensity m a = Codensity (forall b. (a -> m b) -> m b)

toCod :: Monad m => m a -> Codensity m a
toCod ma = Codensity (ma >>=)

fromCod :: Monad m => Codensity m a -> m a
fromCod (Codensity r) = r return

-- round-trip через Codensity (= Ran Maybe Maybe): Just 7
fromCod (toCod (Just 7)) :: Maybe Int
```

<a id="dual-limcolim"></a>
## Пределы / Копределы через Кан

**Идея.** Дана диаграмма `D : J → C` (форма `J`). **Предел** — её универсальный *конус*, **копредел** — универсальный *коконус*. Через движок Кана это в точности расширения вдоль единственного `! : J → 1`:
- `colim D = Lan_! D` (левое — коконус, единица `η`);
- `lim D = Ran_! D` (правое — конус, коединица `ε`).

**Категорный уровень** (расширение вдоль `!`):

![limcolim fig.1](../diagrams/duality/limcolim_1.svg)

Значение расширения в единственном объекте категории `1` — это и есть вершина `colim D` / `lim D`, а единица/коединица `η`/`ε` разворачиваются в **универсальный (ко)конус** на уровне объектов:

![limcolim fig.2](../diagrams/duality/limcolim_2.svg)

**Универсальное свойство.** Любой конус над `D` единственным образом пропускается через `lim D`; любой коконус — через `colim D`. Частные формы `J` (терминальный/начальный, произведение/копроизведение, уравнитель, pullback) — отдельные разделы ниже.

**В чём двойственность:** конус ↔ коконус, `lim ↔ colim` — обращение всех стрелок (`C^op`); на уровне Кана это `Lan ↔ Ran`.

**Где в курсе:** [KanExtensions](KanExtensions.ipynb) — пределы как Кан; [Toposes](Toposes.ipynb) — конкретные конусы (pullback).

```haskell
-- Предел = универсальный конус, копредел = универсальный коконус.
-- На конкретной диаграмме из двух объектов (Int, Bool):
--   предел   = произведение (Int, Bool) с проекциями (конус),
--   копредел = сумма Either Int Bool с инъекциями (коконус).

-- универсальный конус: проекции произведения
prj1 :: (Int, Bool) -> Int
prj1 = fst
prj2 :: (Int, Bool) -> Bool
prj2 = snd

-- любой конус (f, g) пропускается через предел ЕДИНСТВЕННЫМ медиатором <f, g>
mediateCone :: (x -> Int) -> (x -> Bool) -> x -> (Int, Bool)
mediateCone f g x = (f x, g x)

-- двойственно: универсальный коконус -- инъекции суммы, медиатор [f, g]
injL :: Int -> Either Int Bool
injL = Left
injR :: Bool -> Either Int Bool
injR = Right
mediateCocone :: (Int -> y) -> (Bool -> y) -> Either Int Bool -> y
mediateCocone = either

-- демо: медиатор конуса даёт пару; медиатор коконуса разбирает сумму
(mediateCone (+1) (> 0) (5 :: Int), mediateCocone show (\b -> if b then "yes" else "no") (Left 7 :: Either Int Bool))
```

<a id="dual-termini"></a>
## Терминальный / Начальный объект (J = &#8709;)

**Идея.** Простейшая форма — **пустая** диаграмма `J = &#8709;`. Её предел — **терминальный** объект `1` (универсальный конус без ног), копредел — **начальный** объект `0` (универсальный коконус без ног). Через Кан вдоль `! : &#8709; → 1`: `0 = Lan_! &#8709;`, `1 = Ran_! &#8709;`. Двойственность: `0` в `C` — это `1` в `C^op`.

**Категорный уровень** (расширение пустой диаграммы):

![termini fig.1](../diagrams/duality/termini_1.svg)

**Объектный уровень** — что это значит в `C`: начальный `0` шлёт **единственную** стрелку в любой объект, терминальный `1` принимает единственную из любого объекта (обращение стрелок).

![termini fig.2](../diagrams/duality/termini_2.svg)

**В чём двойственность:** начальный объект `0` — это терминальный объект `1` в `C^op` (обращаем единственные стрелки).

**Где в курсе:** [TypeAlgebra](TypeAlgebra.ipynb) — `Void` (= `0`) и `()` (= `1`), кардинальности типов.

```haskell
import Data.Void (Void, absurd)

-- Начальный объект 0 = Void: ЕДИНСТВЕННАЯ стрелка в любой тип a.
fromInitial :: Void -> a
fromInitial = absurd

-- Терминальный объект 1 = (): ЕДИНСТВЕННАЯ стрелка из любого типа a.
toTerminal :: a -> ()
toTerminal _ = ()

-- демо единственности: куда ни шли -- результат один и тот же ()
(toTerminal (42 :: Int), toTerminal "hi")
```

## Произведение / Копроизведение (J = пара)

**Идея.** Форма `J` — **дискретная пара** (два объекта, без стрелок между ними). Её предел — **произведение** `a×b`, копредел — **копроизведение** `a+b`. Через Кан вдоль `! : J → 1`: `a+b = Lan_! D`, `a×b = Ran_! D`. Двойственность: произведение в `C` — это копроизведение в `C^op`.

**Категорный уровень** (форма `J = • •`):

![prodcoprod fig.1](../diagrams/duality/prodcoprod_1.svg)

**Объектный уровень — универсальное свойство.** Произведение `a×b` — универсальный конус: проекции `π1, π2`, и любой `x` с `f : x→a`, `g : x→b` пропускается **единственным** `⟨f,g⟩ : x → a×b`. Копроизведение `a+b` дуально: инъекции `ι1, ι2`, единственный `[f,g] : a+b → x`. (`ι/π` — это компоненты тех самых `η/ε` универсального (ко)конуса из раздела «Пределы/Копределы».)

![prodcoprod fig.2](../diagrams/duality/prodcoprod_2.svg)

**В чём двойственность:** копроизведение `a+b` — это произведение в `C^op` (инъекции `ι` ↔ проекции `π`, медиатор `[f,g]` ↔ `⟨f,g⟩`).

**Где в курсе:** [TypeAlgebra](TypeAlgebra.ipynb) — `(a,b)` (= `a×b`) и `Either a b` (= `a+b`), `Either=+`, `(,)=×`.

```haskell
-- Произведение a*b = (a, b): проекции fst/snd, медиатор (&&&) = <f,g>.
-- Копроизведение a+b = Either a b: инъекции Left/Right, медиатор either = [f,g].

-- product: для f : x -> a и g : x -> b существует ЕДИНСТВЕННЫЙ <f,g> : x -> (a,b)
mediateProd :: (x -> a) -> (x -> b) -> x -> (a, b)
mediateProd f g z = (f z, g z)

-- coproduct (двойственно): для f : a -> x и g : b -> x существует [f,g] : Either a b -> x
mediateCoprod :: (a -> x) -> (b -> x) -> Either a b -> x
mediateCoprod = either

demoProd   = mediateProd (+1) even (10 :: Int)
demoCoprod = mediateCoprod show (\b -> if b then "Y" else "N") (Right True :: Either Int Bool)

(demoProd, demoCoprod)
```

## Уравнитель / Коуравнитель (J = параллельная пара)

**Идея.** Форма `J` — **параллельная пара** `• ⇉ •` (два объекта и две стрелки `f, g` между ними). Её предел — **уравнитель** `eq(f,g)`: универсальный подобъект `e ↣ a`, на котором `f` и `g` совпадают. Копредел — **коуравнитель** `coeq(f,g)`: универсальный фактор `b ↠ q`, склеивающий `f` и `g`. Через Кан вдоль `! : J → 1`: `coeq = Lan_! D`, `eq = Ran_! D`. Двойственность: уравнитель в `C` — это коуравнитель в `C^op`.

**Категорный уровень** (форма `J = • ⇉ •`):

![equ fig.1](../diagrams/duality/equ_1.svg)

**Объектный уровень — универсальное свойство.** Уравнитель `e --eq--> a` удовлетворяет `f&#8728;eq = g&#8728;eq`, и любой `h : x → a` с `f&#8728;h = g&#8728;h` пропускается **единственным** `u : x → e` (`eq&#8728;u = h`). Коуравнитель `b --coeq--> q` дуально: `coeq&#8728;f = coeq&#8728;g`, и любой `k : b → x` с `k&#8728;f = k&#8728;g` пропускается единственным `u : q → x`.

![equ fig.2](../diagrams/duality/equ_2.svg)

**В чём двойственность:** коуравнитель `coeq(f,g)` — это уравнитель в `C^op` (подобъект `e ↣ a` ↔ фактор `b ↠ q`, медиатор `u : x → e` ↔ `u : q → x`).

**Где в курсе:** [KanExtensions](KanExtensions.ipynb) — пределы как Кан; [Toposes](Toposes.ipynb) — подобъекты и классификатор `&#937;` (уравнитель `↣` как характеристика).

```haskell
-- Уравнитель f,g : a -> b в Set: подобъект e = { x | f x = g x } (включение eq : e -> a).
-- Любой h : x -> a с f.h = g.h пропускается единственным u : x -> e (eq.u = h).
equalizer :: Eq b => (a -> b) -> (a -> b) -> [a] -> [a]
equalizer f g = filter (\x -> f x == g x)

-- Коуравнитель (двойственно): фактор q объекта b, склеивающий f x ~ g x (coeq : b -> q).
-- Любой k : b -> x с k.f = k.g пропускается единственным u : q -> x.
coequalizer :: Eq b => (a -> b) -> (a -> b) -> [a] -> b -> b
coequalizer f g pts = \y -> foldr glue y [ (f x, g x) | x <- pts ]
  where glue (u, v) z = if z == v then u else z   -- представители v склеиваем в u

demoEqu   = equalizer (`mod` 2) (`mod` 3) [0 .. 6 :: Int]
demoCoequ = map (coequalizer (const True) (const False) [()]) [True, False]

(demoEqu, demoCoequ)
```

## Pullback / Pushout (J = кослияние / слияние)

**Идея.** Форма `J` — **кослияние** (cospan) `• → • ← •` для предела и **слияние** (span) `• ← • → •` для копредела. Предел кослияния `a →f c ←g b` — **pullback** `a ×_c b` (универсальный квадрат с проекциями `p1, p2` и `f&#8728;p1 = g&#8728;p2`). Копредел слияния `b ←f a →g c` — **pushout** (универсальный квадрат с инъекциями `i1, i2`, склеивающий `f` и `g`). Через Кан вдоль `! : J → 1`: `pushout = Lan_! D`, `pullback = Ran_! D`. Двойственность: pullback в `C` — это pushout в `C^op`.

**Категорный уровень** (формы `J`):

![pbpo fig.1](../diagrams/duality/pbpo_1.svg)

**Объектный уровень — универсальное свойство.** Pullback `a ×_c b` — универсальный квадрат: проекции `p1, p2` с `f&#8728;p1 = g&#8728;p2`, и любой `x` с `q1 : x→a`, `q2 : x→b`, `f&#8728;q1 = g&#8728;q2` пропускается **единственным** `u : x → a×_c b`. Pushout дуально: инъекции `i1, i2`, любой согласованный `(k1, k2)` пропускается единственным `u : po → x`.

![pbpo fig.2](../diagrams/duality/pbpo_2.svg)

**В чём двойственность:** pushout `po` — это pullback в `C^op` (инъекции `i1, i2` ↔ проекции `p1, p2`, медиатор `u : po → x` ↔ `u : x → a×_c b`, склейка ↔ согласованность).

**Где в курсе:** [Toposes](Toposes.ipynb) — pullback (прообраз, характеристическая стрелка вдоль `&#937;`); [KanExtensions](KanExtensions.ipynb) — pullback/pushout как Кан.

```haskell
-- Pullback a x_c b = { (x,y) | f x = g y } : предел кослияния  a --f--> c <--g-- b.
-- Проекции p1 = fst, p2 = snd; любой согласованный (q1,q2) пропускается единственным u.
pullback :: Eq c => (a -> c) -> (b -> c) -> [a] -> [b] -> [(a, b)]
pullback f g as bs = [ (x, y) | x <- as, y <- bs, f x == g y ]

-- Pushout (двойственно): копредел слияния  b <--f-- a --g--> c.
-- Фактор b + c (инъекции i1 = Left, i2 = Right), склеивающий f s ~ g s для s из a.
pushout :: Eq c => (a -> b) -> (a -> c) -> [a] -> Either b c -> Either b c
pushout f g pts = norm
  where norm (Right y) = case [ f s | s <- pts, g s == y ] of
                           (b : _) -> Left b      -- склеиваем g s в f s
                           []      -> Right y
        norm e         = e

demoPB = pullback (`mod` 2) (`mod` 2) [1, 2, 3 :: Int] [4, 5 :: Int]
demoPO = map (pushout (const (1 :: Int)) (const 'x') [()]) [Left 1, Right 'x', Right 'y']

(demoPB, demoPO)
```

<a id="dual-adj"></a>
## Сопряжения через Кан (F ⊣ G)

**Идея.** Сопряжение `F ⊣ G` (`F : C → D` слева, `G : D → C` справа) — двойственная пара, выраженная **через движок Кана**: правый сопряжённый — это правое расширение тождества вдоль левого, левый — левое расширение тождества вдоль правого:
- `G = Ran_F Id_C` (коединица `ε : G&#8728;F ⇒ Id`),
- `F = Lan_G Id_D` (единица `η : Id ⇒ F&#8728;G`).

Биекция `Hom_D(F a, b) ≅ Hom_C(a, G b)` — это универсальное свойство; в Haskell — `curry`/`uncurry` сопряжения `(−,s) ⊣ (s→)`.

**Категорный уровень** (сопряжённый = расширение Кана тождества):

![adj fig.1](../diagrams/duality/adj_1.svg)

**Объектный уровень — универсальная стрелка.** Единица `η_a : a → G(F a)` **универсальна**: для любого `b` и `f : a → G b` существует **единственный** `g : F a → b` с `G g &#8728; η_a = f`. Коединица `ε_b : F(G b) → b` couniversal дуально: любой `g : F a → b` даёт единственный `f : a → G b` с `ε_b &#8728; F f = g`.

![adj fig.2](../diagrams/duality/adj_2.svg)

**В чём двойственность:** единица `η` ↔ коединица `ε`, левый `F` ↔ правый `G` — обращение стрелок (`C^op`); на уровне Кана `Lan_G Id ↔ Ran_F Id`. Сопряжение в `C` ↔ сопряжение `G^op ⊣ F^op` в `C^op`.

**Где в курсе:** [Adjunctions](Adjunctions.ipynb) — `curry`/`uncurry`, единица/коединица, треугольные тождества; [KanExtensions](KanExtensions.ipynb) — сопряжённый как `Ran_F Id` / `Lan_G Id`.

```haskell
-- Сопряжение (-, s) -| (s ->) : левый F a = (a, s), правый G b = s -> b.
-- Биекция Hom_D(F a, b) ~ Hom_C(a, G b) — это в точности curry / uncurry.
leftAdj :: ((a, s) -> b) -> (a -> s -> b)    -- транспонирование влево  = curry
leftAdj = curry

rightAdj :: (a -> s -> b) -> ((a, s) -> b)   -- транспонирование вправо = uncurry
rightAdj = uncurry

eta :: a -> s -> (a, s)        -- единица   eta : a -> G (F a)   (универсальная стрелка)
eta a s = (a, s)

eps :: (s -> b, s) -> b        -- коединица eps : F (G b) -> b   (вычисление, couniversal)
eps (f, s) = f s

demoAdj = (leftAdj (\(n, s) -> n + length s) 10 "abc", rightAdj (\n s -> n + length s) (10, "abc"))
demoTri = (eta (7 :: Int) "xy", eps (length, "abcd"))

(demoAdj, demoTri)
```

## Монада / Комонада через Кан (f вдоль себя)

**Идея.** Финал «всё — расширение Кана»: продолжаем функтор `f : C → D` **вдоль самого себя** и получаем эндофунктор на `D` с канонической (ко)монадной структурой:
- `T = Ran_f f` — **монада коплотности** (codensity): единица `η : Id ⇒ T` (`return`), умножение `μ : T&#178; ⇒ T` (`join`);
- `D = Lan_f f` — **комонада плотности** (density): коединица `ε : D ⇒ Id` (`extract`), коумножение `δ : D ⇒ D&#178;` (`duplicate`).

Замечательно: `Codensity m` — монада для **любого** `m` (даже не-монады), `Density w` — комонада для любого `w`. Двойственность `Lan ↔ Ran` здесь — это `комонада ↔ монада`.

**Категорный уровень** (расширение `f` вдоль `f`):

![monadcomonad fig.1](../diagrams/duality/monadcomonad_1.svg)

**Уровень операций — (ко)единица и (ко)умножение.** Монада `T` собирает структуру в `T a` (стрелки **входят**: `η : a → T a`, `μ : T&#178; a → T a`); комонада `D` раздаёт из `D a` (стрелки **выходят**: `ε : D a → a`, `δ : D a → D&#178; a`). Это буквальное обращение стрелок.

![monadcomonad fig.2](../diagrams/duality/monadcomonad_2.svg)

**В чём двойственность:** `Lan_f f` (комонада, `ε`/`δ`) ↔ `Ran_f f` (монада, `η`/`μ`) — обращение всех стрелок (`C^op`). Комонада в `C` — это монада в `C^op`.

**Где в курсе:** [KanExtensions](KanExtensions.ipynb) — `Codensity = Ran f f`, плотность/коплотность; [Monads](Monads.ipynb) — `return`/`join`; комонады — `extract`/`duplicate`.

```haskell
-- Codensity m = Ran m m : МОНАДА на любом функторе m (даже если m — не монада!).
newtype Codensity m a = Codensity { runCodensity :: forall b. (a -> m b) -> m b }

instance Functor (Codensity m) where
  fmap f (Codensity g) = Codensity (\k -> g (k . f))
instance Applicative (Codensity m) where
  pure a = Codensity (\k -> k a)                       -- единица  eta : a -> T a
  Codensity f <*> Codensity g = Codensity (\k -> f (\h -> g (k . h)))
instance Monad (Codensity m) where
  Codensity g >>= f = Codensity (\k -> g (\a -> runCodensity (f a) k))   -- mu : T T a -> T a

toCodensity :: Monad m => m a -> Codensity m a
toCodensity m = Codensity (m >>=)
fromCodensity :: Monad m => Codensity m a -> m a
fromCodensity (Codensity g) = g return

-- Density w = Lan w w : КОМОНАДА на любом функторе w (двойственно).
data Density w a = forall b. Density (w b -> a) (w b)

instance Functor (Density w) where
  fmap f (Density g w) = Density (f . g) w
extractD :: Density w a -> a                            -- коединица eps : D a -> a (extract)
extractD (Density f w) = f w
duplicateD :: Density w a -> Density w (Density w a)     -- delta : D a -> D D a
duplicateD (Density f w) = Density (Density f) w

demoCod = fromCodensity (do { x <- toCodensity (Just 7); toCodensity (Just (x + 1)) }) :: Maybe Int
demoDen = ( extractD (Density sum [1, 2, 3 :: Int])
          , extractD (extractD (duplicateD (Density sum [1, 2, 3 :: Int]))) )

(demoCod, demoDen)
```
