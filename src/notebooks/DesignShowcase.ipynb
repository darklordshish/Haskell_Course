Ноутбуки имели **8 точек несогласованности**:

| Аспект | Разнобой | Решение |
|--------|----------|---------|

| **Оглавление** | Таблица/список/нет | Таблица 📌 везде |
| **Номера секций** | `## 1.` vs `## 1️⃣` | `## 1️⃣` (с эмодзи) |
| **Категорный блок** | `### 🔺` только кое-где | В каждой секции |
| **Setup-ячейка** | Есть/нет | Всегда первая |
| **Разделители** | Непоследовательно | `---` между каждой секцией |
| **Навигация** | В конце/начале/нет | Всегда последняя ячейка |
| **SVG палитра** | 5+ разных цветовых схем | Единая светлая (в тон карте) |

## 🇗️ Структура ноутбука (шаблон)

```
[1] SETUP-ЯЧЕЙКА (code cell, всегда первая)
    :set -XExtensions
    import ...
    putStrLn "Setup done."

[2] ЗАГОЛОВОК (markdown)
    # 🗺️ Тема в Haskell
    Одно-два предложения — суть.

[3] ОГЛАВЛЕНИЕ (markdown)
    📌 Содержание
    | # | Тема | Суть |
    |---|------|------|
    | 1 | ...  | ...  |

---

[4..N] СЕКЦИИ (markdown + code чередуются)
    ## 1️⃣ Раздел
    Определение + законы.
    | Закон | Формула |

    ### 🔺 Категорный взгляд
    Категорная интерпретация.

    ![описание](filename.svg)

    [code cell]
---

[N+1] ИТОГ (markdown)
    ## 📊 Итог
    Сравнительная таблица / резюме.

[N+2] НАВИГАЦИЯ (markdown, последняя)
    ---
    **← Предыдущий:** [X](X.ipynb) | **→ Следующий:** [Y](Y.ipynb)
```

## ➡️ Кит маркеров стрелок (копировать в `<defs>` нового SVG)

```xml
<defs>
  <marker id="ah" markerWidth="10" markerHeight="8" refX="8" refY="3.5" orient="auto" markerUnits="userSpaceOnUse"><path d="M0,0 L9,3.5 L0,7 z" fill="#8a93a3"/></marker>
  <marker id="ah-epi" markerWidth="15" markerHeight="8" refX="13" refY="3.5" orient="auto" markerUnits="userSpaceOnUse"><path d="M0,0 L9,3.5 L0,7 z M5,0 L14,3.5 L5,7 z" fill="#8a93a3"/></marker>
  <marker id="ah-tail" markerWidth="10" markerHeight="8" refX="1" refY="3.5" orient="auto" markerUnits="userSpaceOnUse"><path d="M8,0 L1,3.5 L8,7" fill="none" stroke="#8a93a3" stroke-width="1.6"/></marker>
</defs>
```

| Морфизм | Как рисовать |
|---------|--------------|
| `→` общий | `marker-end="url(#ah)"` |
| `↣` mono (моно) | `marker-start="url(#ah-tail)" marker-end="url(#ah)"` |
| `↠` epi (эпи) | `marker-end="url(#ah-epi)"` |
| `≅` iso | `marker-end="url(#ah)"` + текст `≅` над линией |
| `⇒` nat (ест. преобр.) | две параллельные линии + `marker-end="url(#ah)"` |

Наконечники по умолчанию нейтральные `#8a93a3`; цветной — перекрашенная копия маркера.
Применять mono/epi/iso/nat только там, где морфизм действительно таков.

## 1️⃣ Functor — эндофунктор в Hask

Функтор `f :: * -> *` — это эндофунктор категории **Hask**: отображает типы в типы и функции в функции.

```haskell
class Functor f where
    fmap :: (a -> b) -> f a -> f b
```

| Закон | Формула | Суть |
|-------|---------|------|
| Тождество | `fmap id = id` | не меняет структуру |
| Композиция | `fmap (f . g) = fmap f . fmap g` | сохраняет цепочки |

### 🔺 Категорный взгляд

Функтор `F : C → D` — это пара отображений:  
- на **объектах**: `F(A)` для каждого объекта `A ∈ C`  
- на **морфизмах**: `F(f) : F(A) → F(B)` для каждого `f : A → B`

Два закона = `F` сохраняет тождественные стрелки и композицию.
