## 📊 Анализ — что было не так

Ноутбуки имели **8 точек несогласованности**:

| Аспект | Разнобой | Решение |
|--------|----------|---------|
| **Заголовки H1** | С эмодзи/без | `# 🗺️ Тема в Haskell` |
| **Оглавление** | Таблица/список/нет | Таблица 📌 везде |
| **Номера секций** | `## 1.` vs `## 1️⃣` | `## 1️⃣` (с эмодзи) |
| **Категорный блок** | `### 🔺` только кое-где | В каждой секции |
| **Setup-ячейка** | Есть/нет | Всегда первая |
| **Разделители** | Непоследовательно | `---` между каждой секцией |
| **Навигация** | В конце/начале/нет | Всегда последняя ячейка |
| **SVG палитра** | 5+ разных цветовых схем | Единая тёмная |

## 🇗️ Структура ноутбука (шаблон)

```
[1] SETUP-ЯЧЕЙКА (code cell, всегда первая)
    :set -XExtensions
    import ...
    putStrLn "Setup done."

[2] ЗАГОЛОВОК (markdown)
    # 🗺️ Тема в Haskell
    Одно-два предложения — суть.

[3] ОГЛАВЛЕНИЕ (markdown)
    📌 Содержание
    | # | Тема | Суть |
    |---|------|------|
    | 1 | ...  | ...  |

---

[4..N] СЕКЦИИ (markdown + code чередуются)
    ## 1️⃣ Раздел
    Определение + законы.
    | Закон | Формула |

    ### 🔺 Категорный взгляд
    Категорная интерпретация.

    ![описание](filename.svg)

    [code cell]
---

[N+1] ИТОГ (markdown)
    ## 📊 Итог
    Сравнительная таблица / резюме.

[N+2] НАВИГАЦИЯ (markdown, последняя)
    ---
    **← Предыдущий:** [X](X.ipynb) | **→ Следующий:** [Y](Y.ipynb)
```

## 📚 Правила заголовков

| Уровень | Формат | Пример |
|---------|--------|--------|
| **H1** (название) | `# [эмодзи] [Тема] в Haskell` | `# 🗺️ Монады в Haskell` |
| **H2** (секция) | `## N️⃣ Название` | `## 1️⃣ Functor` |
| **H3** категорный | `### 🔺 Категорный взгляд` | всегда именно так |
| **H3** реализация | `### 📐 Реализация` | разбор кода |
| **H3** примеры | `### 💡 Примеры` | практические примеры |
| **H3** итог | `### 📊 Итог / Сравнение` | итоговая таблица |

**Нумерация H2:** 1️⃣ 2️⃣ 3️⃣ 4️⃣ 5️⃣ 6️⃣ 7️⃣ 8️⃣ 9️⃣ 🔟

## 1️⃣ Functor — эндофунктор в Hask

Функтор `f :: * -> *` — это эндофунктор категории **Hask**: отображает типы в типы и функции в функции.

```haskell
class Functor f where
    fmap :: (a -> b) -> f a -> f b
```

| Закон | Формула | Суть |
|-------|---------|------|
| Тождество | `fmap id = id` | не меняет структуру |
| Композиция | `fmap (f . g) = fmap f . fmap g` | сохраняет цепочки |

### 🔺 Категорный взгляд

Функтор `F : C → D` — это пара отображений:  
- на **объектах**: `F(A)` для каждого объекта `A ∈ C`  
- на **морфизмах**: `F(f) : F(A) → F(B)` для каждого `f : A → B`

Два закона = `F` сохраняет тождественные стрелки и композицию.
