# 🪞 Категория `Set^op`

Один странный объект — три взгляда. Всё в сравнении с `Set`.

## 📌 Содержание

<a id="setop-intro"></a>
## Set^op: переверни стрелки

**Парадокс.** Двойственность кажется тривиальной: берём категорию, переворачиваем все стрелки — готова `C^op`. Но `Set` — самая интуитивная категория (множества и функции), а что такое `Set^op`? Определяется в одну строку, а представить его как «объект» неожиданно трудно: угадал все буквы, не складывается в слово. Этот этюд разглядывает `Set^op` с трёх сторон — что это (ветка **A**), есть ли там логика (**B**), где применяется (**C**) — всё время держа рядом `Set`.

**Определение.** В `C^op` — те же объекты, а `Hom_{op}(A,B) = Hom(B,A)`, композиция в обратном порядке. Конкретно для множеств: **стрелка `A → B` в `Set^op` — это функция `B → A` в `Set`**. Объекты те же; меняется только направление стрелок.

![intro fig.1](../diagrams/setop/intro_1.svg)

Дальше — три ветки из этого определения.

<a id="setop-limits"></a>
## (Ко)пределы и mono/epi в Set^op

Раз `Hom_{op}(A,B) = Hom(B,A)`, любая универсальная конструкция в `Set^op` — это двойственная конструкция в `Set` (полный словарь — [Duality.ipynb](Duality.ipynb); здесь — конкретика для множеств):

- **Mono / epi.** Мономорфизм в `Set` = инъекция, эпиморфизм = сюръекция. В `Set^op` они меняются ролями: мономорфизм `Set^op` приходит из сюръекции `Set`, эпиморфизм — из инъекции. Поэтому **субобъекты `Set^op` — это сюръекции/факторы `Set`** (важно для ветки B).
- **Терминал / начальный.** Терминальный объект `Set` — синглтон `{∗}`, начальный — `∅`. В `Set^op` они переворачиваются: **терминальный объект `Set^op` — это `∅`** (!), начальный — `{∗}`.
- **(Ко)пределы.** Произведение `Set^op` = копроизведение `Set` = дизъюнктное объединение `⊔`; копроизведение `Set^op` = произведение `Set` = `×`. `Set` полна и кополна ⇒ `Set^op` тоже — пределы и копределы просто меняются местами.

![limits fig.1](../diagrams/setop/limits_1.svg)

Структура есть, но «привычной» `Set^op` быть перестаёт — следующий раздел показывает, чего у него нет.

<a id="setop-notccc"></a>
## Не декартово замкнута, не топос

У `Set` есть экспоненты `B^A` (множество функций) — это делает её **декартово замкнутой** (CCC), и именно поэтому в ней так удобно «программировать». А `Set^op`?

**Нет объектов-функций.** `Set^op` была бы CCC ⟺ в `Set` существовали бы **ко-экспоненты** (объекты, двойственные экспонентам). Но в `Set` ко-экспонентов нет (кроме вырожденных случаев): конструкция `B^A` не переворачивается. Значит `Set^op` **не декартово замкнута** — в ней нет внутренних объектов-функций. Это первый крупный подводный камень: интуиция «функции между множествами» при переходе в `Set^op` теряет носитель.

**Параллель в Haskell.** `Hask` (типы и функции) — декартово замкнута: тип `a -> b` — это объект-экспонента, на нём держится всё каррирование. А `Hask^op` декартово замкнутой не является — поэтому «программировать в противоположной категории» напрямую нельзя; её тень — это CPS/континуации (ветка **C**).

**Значит, и не топос.** Элементарный топос обязан быть декартово замкнутым; раз `Set^op` не CCC, она **не топос** — у неё нет классификатора подобъектов `Ω` в смысле топосов. Но какая-то структура подобъектов у неё всё же есть. Вопрос «а что там вместо логики топоса?» — ровно тема ветки **B**, к которой мы и переходим.

<a id="setop-caba"></a>
## A. Что это: Set^op ≃ CABA

Первый взгляд — отождествить `Set^op` с чем-то знакомым.

**Powerset контравариантен.** Сопоставим множеству `X` булеву алгебру подмножеств `2^X`. Функция `f : A → B` даёт **прообраз** `f⁻¹ : 2^B → 2^A`, где `f⁻¹(S) = { x | f x ∈ S }` — стрелка в обратную сторону. Прообраз сохраняет `∪`, `∩` и дополнение, то есть это гомоморфизм булевых алгебр. Так powerset — контравариантный функтор `P : Set^op → CABA`.

![caba fig.1](../diagrams/setop/caba_1.svg)

**Теорема: `Set^op ≃ CABA`.** Категория `Set^op` **эквивалентна** категории полных атомарных булевых алгебр (CABA) с полными гомоморфизмами. Множество `X` восстанавливается из `2^X` как множество его **атомов** — одноэлементных подмножеств `{x}`. Значит «объект `Set^op`» — это, по сути, **булева алгебра его подмножеств**, а элементы `X` — атомы этой алгебры.

![caba fig.2](../diagrams/setop/caba_2.svg)

**Стрелка = предикат-трансформер.** Раз стрелка `A → B` в `Set^op` — это `f : B → A`, на предикатах она действует как `f⁻¹`: тащит предикат на `A` в предикат на `B` против направления (weakest precondition). В Haskell — контравариантный функтор `Op r a = a → r`, где `contramap` подставляет `f` перед предикатом.

```haskell
-- Contravariant powerset: f : a -> b yields preimage f_inv : 2^b -> 2^a.
-- f_inv is a boolean algebra homomorphism (preserves intersection/union/complement).
powerset :: [a] -> [[a]]
powerset = subsequences

preimage :: Eq b => (a -> b) -> [a] -> [b] -> [a]
preimage fn dom s = [ x | x <- dom, fn x `elem` s ]

setA :: [Int]
setA = [0 .. 5]

fMod :: Int -> Int
fMod = (`mod` 3)                   -- f : A -> B = {0,1,2}

inter :: Eq a => [a] -> [a] -> [a]
inter xs ys = [x | x <- xs, x `elem` ys]

union' :: Eq a => [a] -> [a] -> [a]
union' xs ys = nub (xs ++ ys)

s1 :: [Int]
s1 = [0,1]
s2 :: [Int]
s2 = [1,2]

checkInter :: Bool
checkInter = sort (preimage fMod setA (inter s1 s2)) == sort (inter (preimage fMod setA s1) (preimage fMod setA s2))

checkUnion :: Bool
checkUnion = sort (preimage fMod setA (union' s1 s2)) == sort (union' (preimage fMod setA s1) (preimage fMod setA s2))

newtype Op r a = Op { runOp :: a -> r }

contramapOp :: (a -> b) -> Op r b -> Op r a
contramapOp g (Op h) = Op (h . g)

adult :: Op Bool Int
adult = Op (>= (18 :: Int))

byLen :: Op Bool String
byLen = contramapOp length adult

demoPre :: [Int]
demoPre = preimage fMod setA [1]

demoA :: Bool
demoA = runOp adult 20

demoB :: Bool
demoB = runOp byLen "hi"

(demoPre, checkInter, checkUnion, demoA, demoB)
```

<a id="setop-partitions"></a>
## B1. Субобъекты = разбиения (логики нет)

В топосе `Set` логика живёт в решётках субобъектов: `Sub(X) = 2^X` — булева алгебра, классическая логика с `Ω = {0,1}`. А что в `Set^op`?

**Субобъект `Set^op` = фактор `Set` = разбиение.** Мономорфизмы `Set^op` приходят из эпиморфизмов `Set` (см. раздел про mono/epi) — то есть из **сюръекций** `X ↠ Q`. Сюръекция с точностью до изоморфизма — это **разбиение** `X` на блоки (слои). Значит «решётка субобъектов» объекта `X` в `Set^op` — это решётка разбиений `Π(X)`.

**И вот сюрприз: эта решётка недистрибутивна.** Уже для трёхэлементного множества `Π(3) = M₃` — пятиэлементная «диамант»-решётка, нарушающая дистрибутивность; для `Π(4)` теряется и модулярность. Ни булевой, ни Гейтинговой структуры здесь нет — значит, нет и пропозициональной логики субобъектов. Наивная надежда «двойственность `Set` даст красивую co-логику» проваливается прямо тут.

![partitions fig.1](../diagrams/setop/partitions_1.svg)

Код ниже строит решётку разбиений `{1,2,3}` и предъявляет контрпример дистрибутивности.

```haskell
type Partition = [[Int]]

norm :: Partition -> Partition
norm = sort . map sort

allPartitions :: [Int] -> [Partition]
allPartitions [] = [[]]
allPartitions (x:xs) = concatMap ins (allPartitions xs)
  where ins p = ([x] : p) : [ (x : b) : rest | (b, rest) <- picks p ]
        picks [] = []
        picks (b:bs) = (b, bs) : [ (b', b : bs') | (b', bs') <- picks bs ]

meetP :: Partition -> Partition -> Partition
meetP a b = norm [ i | ba <- a, bb <- b, let i = [ x | x <- ba, x `elem` bb ], not (null i) ]

joinP :: Partition -> Partition -> Partition
joinP a b = norm (merge (nub (a ++ b)))
  where merge bs = case [ (x, y) | x <- bs, y <- bs, x < y, any (`elem` y) x ] of
                     []        -> bs
                     ((x,y):_) -> merge (nub (x ++ y) : filter (\z -> z /= x && z /= y) bs)

pa = [[1,2],[3]] :: Partition
pb = [[1,3],[2]] :: Partition
pc = [[2,3],[1]] :: Partition
lhs = meetP pa (joinP pb pc)
rhs = joinP (meetP pa pb) (meetP pa pc)

(length (allPartitions [1,2,3]), norm lhs, norm rhs, norm lhs == norm rhs)
```

<a id="setop-coheyting"></a>
## B2-3. Ко-Гейтинг, граница, двойное отрицание

Если прямой `Set^op` логики не дал (B1), то **настоящая** двойственная логика появляется иначе — переворотом порядка Гейтинговой алгебры.

**Ко-Гейтинг.** Алгебра Гейтинга `H` (например, `Sub(X)` в топосе) несёт импликацию `→` (правый сопряжённый к `∧`) и отрицание `¬a`, причём `a ∧ ¬a = ⊥`. Перевёрнутая `H^op` — **ко-Гейтингова** (брауэрова) алгебра: вместо импликации — **вычитание** `a − b` (левый сопряжённый к `∨`), ко-отрицание `~a = ⊤ − a`, и **граница** `∂a = a ∧ ~a`. Ключевое отличие: `∂a` может быть `≠ ⊥` — это **паранепротиворечивость** (на границе `a` и `~a` держатся вместе).

**Где это живёт буквально.** В топологическом пространстве **открытые** множества образуют алгебру Гейтинга (локаль), а **замкнутые** — ко-Гейтингову, и `∂a` — это в точности **топологическая граница**.

![coheyting fig.1](../diagrams/setop/coheyting_1.svg)

![coheyting fig.2](../diagrams/setop/coheyting_2.svg)

**Мост двойного отрицания (связка A↔B↔C).** В `Set` всё булево: Гейтинг = ко-Гейтинг = булева алгебра, граница пуста. Закрытие `¬¬` (нуклеус) схлопывает интуиционистскую логику в булеву — это буализация, дающая `CABA` (ветка A). А вычислительная тень `¬¬` — монада двойного отрицания `Cont` при `r = 2` (ветка C). Так `Set^op ≃ CABA` оказывается «схлопнутым `¬¬`-краем», где паранепротиворечивость вырождается; живая граница `∂a ≠ ⊥` требует не-булева топоса.

**Где в курсе:** [Toposes.ipynb](Toposes.ipynb) — алгебра Гейтинга субобъектов (Голдблатт); Lawvere–Reyes — ко-Гейтинг и граница; Mortensen — паранепротиворечивая математика.

Код ниже: топология цепи `1<2<3`; на открытых — Гейтинг (`¬a`, `a∧¬a = ∅`), на замкнутых — ко-Гейтинг (`~a`, граница `∂a ≠ ∅`).

```haskell
-- Конечная топология: открытые = up-sets цепи 1<2<3 (Alexandrov).
-- Открытые дают алгебру Гейтинга, замкнутые — ко-Гейтинга.
universe = [1,2,3] :: [Int]
opens = [ [], [3], [2,3], [1,2,3] ] :: [[Int]]

subset :: [Int] -> [Int] -> Bool
subset xs ys = all (`elem` ys) xs

complement :: [Int] -> [Int]
complement s = [ x | x <- universe, x `notElem` s ]

interior :: [Int] -> [Int]
interior s = sort (nub (concat [ o | o <- opens, o `subset` s ]))

closure :: [Int] -> [Int]
closure s = complement (interior (complement s))

hImpl :: [Int] -> [Int] -> [Int]
hImpl a b = interior (nub (complement a ++ b))

hNot :: [Int] -> [Int]
hNot a = hImpl a []

cSub :: [Int] -> [Int] -> [Int]
cSub a b = closure [ x | x <- a, x `notElem` b ]

cNot :: [Int] -> [Int]
cNot = cSub universe

boundary :: [Int] -> [Int]
boundary a = [ x | x <- a, x `elem` cNot a ]

oa = [3] :: [Int]
ca = [1,2] :: [Int]
heytNot   = hNot oa
heytMeet  = [ x | x <- oa, x `elem` heytNot ]
coheytNot = cNot ca
coheytBd  = boundary ca

(heytNot, heytMeet, coheytNot, coheytBd)
```

<a id="setop-dualities"></a>
## C1. Пространства против алгебр

Третий взгляд: `Set^op` — частный (вырожденный) случай большой идеи **«геометрия = алгебра наоборот»**.

Снова и снова одна и та же категория «пространств» оказывается **противоположной** категории «алгебр функций на них»:
- **Стоунова двойственность:** булевы алгебры ≃ (Stone-пространства)^op;
- **Гельфанд:** коммутативные C*-алгебры ≃ (компактные хаусдорфовы пространства)^op;
- **Spec:** коммутативные кольца ≃ (аффинные схемы)^op — основа алгебраической геометрии;
- **Локали/фреймы:** фреймы ≃ Locale^op (бесточечная топология).

Общий узор: «пространство» `X` кодируется алгеброй функций на нём, а **стрелки переворачиваются** (отображение пространств ⟶ гомоморфизм алгебр в обратную сторону — как прообраз `f⁻¹` из ветки A). `Set^op ≃ CABA` — тот же узор в самом простом, **дискретном** случае: множество = дискретное пространство, `2^X` = его алгебра «функций в `2`» (характеристические функции = подмножества).

<a id="setop-cps"></a>
## C2-3. CPS-мост: Op, Cont, применения

**Зеркало `Set^op` в Haskell — это CPS.** Контравариантный «функтор предикатов» `Op r a = a → r` тащит наблюдения против стрелок (как `f⁻¹`). Удвоив его, получаем **монаду двойного отрицания** — континуации: `Cont r a = (a → r) → r`. Это вычислительная тень `¬¬`: значение `a` оборачивается в «то, что отдаст `r`, если ему дать продолжение `a → r`». При `r = 2` это буквально двойной powerset (ветка A), при общем `r` — CPS-стиль, где вычисление идёт «наоборот».

![cps fig.1](../diagrams/setop/cps_1.svg)

**Где это всплывает.** CPS-преобразование в компиляторах; delimited continuations (`shift`/`reset`); Stone-двойственность в CS (профинитные моноиды, автоматы — Eilenberg/Stone); интуиция `Spec` в алгебраической геометрии; в базах данных (Спивак) инстансы схемы — функторы, а миграции данных `Σ ⊣ Δ ⊣ Π` живут на стрелках схем.

**Подводные камни.**
- `Set^op ≃ CABA` требует **полноты и атомарности**: для произвольных булевых алгебр получается не `Set`, а Stone-пространства — бесконечный случай теряет «множественность».
- Тип-ответ `r` важен: разные `r` дают разные монады `Cont r`; `r = 2` — булев/классический край, общий `r` — настоящий CPS.
- `Set^op ≠ Set`: в нём **нет** функциональных пространств (не CCC), так что «программировать прямо в `Set^op`» нельзя — только через эту CPS-тень.

Код ниже строит `Op` и монаду `Cont`, показывает единицу двойного отрицания и контравариантный предикат-трансформер.

```haskell
newtype Op r a = Op { runOp :: a -> r }
contramapOp :: (a -> b) -> Op r b -> Op r a
contramapOp f (Op g) = Op (g . f)

newtype Cont r a = Cont { runCont :: (a -> r) -> r }
instance Functor (Cont r) where
  fmap f (Cont c) = Cont (\k -> c (k . f))
instance Applicative (Cont r) where
  pure a = Cont (\k -> k a)
  Cont f <*> Cont x = Cont (\k -> f (\g -> x (k . g)))
instance Monad (Cont r) where
  Cont c >>= f = Cont (\k -> c (\a -> runCont (f a) k))

unit :: a -> Cont r a
unit = pure

adult = Op (>= (18 :: Int)) :: Op Bool Int
byLen = contramapOp length adult :: Op Bool String

demoEval  = runCont (do { x <- unit 7; unit (x + 1) }) id :: Int
demoAdult = runOp adult 20
demoLen   = runOp byLen "hello, world"

(demoEval, demoAdult, demoLen)
```
