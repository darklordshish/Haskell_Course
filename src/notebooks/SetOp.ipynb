<a id="setop-intro"></a>
## Set^op: переверни стрелки

**Парадокс.** Двойственность кажется тривиальной: берём категорию, переворачиваем все стрелки — готова `C^op`. Но `Set` — самая интуитивная категория (множества и функции), а что такое `Set^op`? Определяется в одну строку, а представить его как «объект» неожиданно трудно: угадал все буквы, не складывается в слово. Этот этюд разглядывает `Set^op` с трёх сторон — что это (ветка **A**), есть ли там логика (**B**), где применяется (**C**) — всё время держа рядом `Set`.

**Определение.** В `C^op` — те же объекты, а `Hom_{op}(A,B) = Hom(B,A)`, композиция в обратном порядке. Конкретно для множеств: **стрелка `A → B` в `Set^op` — это функция `B → A` в `Set`**. Объекты те же; меняется только направление стрелок.

![intro fig.1](../diagrams/setop/intro_1.svg)

Дальше — три ветки из этого определения.

<a id="setop-limits"></a>
## (Ко)пределы и mono/epi в Set^op

Раз `Hom_{op}(A,B) = Hom(B,A)`, любая универсальная конструкция в `Set^op` — это двойственная конструкция в `Set` (полный словарь — [Duality.ipynb](Duality.ipynb); здесь — конкретика для множеств):

- **Mono / epi.** Мономорфизм в `Set` = инъекция, эпиморфизм = сюръекция. В `Set^op` они меняются ролями: мономорфизм `Set^op` приходит из сюръекции `Set`, эпиморфизм — из инъекции. Поэтому **субобъекты `Set^op` — это сюръекции/факторы `Set`** (важно для ветки B).
- **Терминал / начальный.** Терминальный объект `Set` — синглтон `{∗}`, начальный — `∅`. В `Set^op` они переворачиваются: **терминальный объект `Set^op` — это `∅`** (!), начальный — `{∗}`.
- **(Ко)пределы.** Произведение `Set^op` = копроизведение `Set` = дизъюнктное объединение `⊔`; копроизведение `Set^op` = произведение `Set` = `×`. `Set` полна и кополна ⇒ `Set^op` тоже — пределы и копределы просто меняются местами.

![limits fig.1](../diagrams/setop/limits_1.svg)

Структура есть, но «привычной» `Set^op` быть перестаёт — следующий раздел показывает, чего у него нет.

<a id="setop-notccc"></a>
## Не декартово замкнута, не топос

У `Set` есть экспоненты `B^A` (множество функций) — это делает её **декартово замкнутой** (CCC), и именно поэтому в ней так удобно «программировать». А `Set^op`?

**Нет объектов-функций.** `Set^op` была бы CCC ⟺ в `Set` существовали бы **ко-экспоненты** (объекты, двойственные экспонентам). Но в `Set` ко-экспонентов нет (кроме вырожденных случаев): конструкция `B^A` не переворачивается. Значит `Set^op` **не декартово замкнута** — в ней нет внутренних объектов-функций. Это первый крупный подводный камень: интуиция «функции между множествами» при переходе в `Set^op` теряет носитель.

**Параллель в Haskell.** `Hask` (типы и функции) — декартово замкнута: тип `a -> b` — это объект-экспонента, на нём держится всё каррирование. А `Hask^op` декартово замкнутой не является — поэтому «программировать в противоположной категории» напрямую нельзя; её тень — это CPS/континуации (ветка **C**).

**Значит, и не топос.** Элементарный топос обязан быть декартово замкнутым; раз `Set^op` не CCC, она **не топос** — у неё нет классификатора подобъектов `Ω` в смысле топосов. Но какая-то структура подобъектов у неё всё же есть. Вопрос «а что там вместо логики топоса?» — ровно тема ветки **B**, к которой мы и переходим.

<a id="setop-caba"></a>
## A. Что это: Set^op ≃ CABA

Первый взгляд — отождествить `Set^op` с чем-то знакомым.

**Powerset контравариантен.** Сопоставим множеству `X` булеву алгебру подмножеств `2^X`. Функция `f : A → B` даёт **прообраз** `f⁻¹ : 2^B → 2^A`, где `f⁻¹(S) = { x | f x ∈ S }` — стрелка в обратную сторону. Прообраз сохраняет `∪`, `∩` и дополнение, то есть это гомоморфизм булевых алгебр. Так powerset — контравариантный функтор `P : Set^op → CABA`.

![caba fig.1](../diagrams/setop/caba_1.svg)

**Теорема: `Set^op ≃ CABA`.** Категория `Set^op` **эквивалентна** категории полных атомарных булевых алгебр (CABA) с полными гомоморфизмами. Множество `X` восстанавливается из `2^X` как множество его **атомов** — одноэлементных подмножеств `{x}`. Значит «объект `Set^op`» — это, по сути, **булева алгебра его подмножеств**, а элементы `X` — атомы этой алгебры.

![caba fig.2](../diagrams/setop/caba_2.svg)

**Стрелка = предикат-трансформер.** Раз стрелка `A → B` в `Set^op` — это `f : B → A`, на предикатах она действует как `f⁻¹`: тащит предикат на `A` в предикат на `B` против направления (weakest precondition). В Haskell — контравариантный функтор `Op r a = a → r`, где `contramap` подставляет `f` перед предикатом.

```haskell
-- Контравариантный powerset: f : a -> b даёт прообраз f^{-1} : 2^b -> 2^a.
-- f^{-1} — гомоморфизм булевых алгебр (сохраняет пересечение/объединение/дополнение).
powerset :: [a] -> [[a]]
powerset = subsequences

preimage :: Eq b => (a -> b) -> [a] -> [b] -> [a]
preimage f dom s = [ x | x <- dom, f x `elem` s ]

setA = [0,1,2,3] :: [Int]
f x = even x                       -- f : A -> B = Bool

inter xs ys = [x | x <- xs, x `elem` ys]
union' xs ys = nub (xs ++ ys)

s1 = [True]; s2 = [False]
checkInter = sort (preimage f setA (inter s1 s2))
           == sort (inter (preimage f setA s1) (preimage f setA s2))
checkUnion = sort (preimage f setA (union' s1 s2))
           == sort (union' (preimage f setA s1) (preimage f setA s2))

-- стрелка Set^op как предикат-трансформер: предикат на b -> предикат на a
newtype Op r a = Op { runOp :: a -> r }
contramapOp :: (a -> b) -> Op r b -> Op r a
contramapOp f (Op g) = Op (g . f)

adult = Op (>= (18 :: Int)) :: Op Bool Int
byLen = contramapOp length adult :: Op Bool String   -- предикат на строках через length

(preimage f setA [True], checkInter, checkUnion, runOp adult 20, runOp byLen "hi")
```
