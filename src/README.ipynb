## 1️⃣1️⃣ 🔭 Расширения Кана

**Файл:** [`KanExtensions.ipynb`](notebooks/KanExtensions.ipynb)  |  **Уровень:** ⚫ эксперт.  |  **Ячеек:** 23  |  **Диаграмм:** 8 SVG  |  **Статус:** ✅ готов

**Предварительно:** [YonedaLemma.ipynb](notebooks/YonedaLemma.ipynb), [MonadTransformers.ipynb](notebooks/MonadTransformers.ipynb)

«Все концепции — это расширения Кана» (МакЛейн).

| Тема | Суть |
|------|------|
| Ran g h | forall b. (a→gb) → hb — правое расширение |
| Lan g h | exists b. (gb→a, hb) — левое расширение |
| Codensity | Ran f f — монада для любого функтора |
| Density | Lan f f — комонада для любого функтора |
| Yoneda через Ran | Ran Id h ≅ h |
| Сопряжения через Ran | Ran f Id ≅ правый сопряжённый |
| Оптимизация | Codensity ускоряет конкатенацию: O(n²) → O(n) |

| Тема | Суть |
