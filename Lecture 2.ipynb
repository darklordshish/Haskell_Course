список некоторых полезных команд ghci:
- :?
     Выводит на экран список доступных команд
 
- :t Expression
     Возвращает тип выражения.
     
- :i 
     Подробная информация о типе, классе типов или функции
     
- :set +t
     После выполнения команды интерпретатор будет выводить на экран не только результат вычисления выражения, но и его тип.
 
- :set +s
     После выполнения команды интерпретатор будет выводить на экран не только результат вычисления выражения, но и статистику вычислений.
 
- :l ИмяМодуля
     Загружает пользовательский модуль в интерпретатор.
     
- :m + ИмяМодуля
     Добавляет внешний модуль в интерпритаторе. 
     
- :cd Директория
     Перейти в данную директорию.
 
- :r
     Перезагружает, последний загруженный модуль. Этой командой можно пользоваться после внесения в модуль изменений.

- :q
     Выход из интерпретатора.

Далее, свои функции будем писать в текстовом редактре, потом загружать в виде модуля. 
Программы на Haskell имеют расширение (.hs). Имена модулей начинаются с заглавной буквы, каждоё слово начинается с заглавной буквы, например "MyFirstFunction.hs"
Чтобы загрузить свой модуль в интерпретатор нужно выполнить команду внутри ghci 
- :l MyFirstFunction.hs

Если Выработаете в jupyter-notebook, 
то можно просто писать в клетке с кодом.

```haskell
-- в файле напишем первую функцию 
doubleMe x = x + x
```

```haskell
doubleMe 3
-- хотя по равилам хорошего тона перед функцией нужно объявить её тип, компилятор сам его выведет
:t doubleMe
```

```haskell
--в файл добавим ещё одну функцию 
doubleUs x y = doubleMe x + doubleMe y
```

```haskell
doubleUs 5 6.7
:t doubleUs
```

```haskell
-- условный оператор 

doubleSmallNumber x = if x > 100  
                        then x  
                        else x*2  
                        
doubleSmallNumber' x = (if x > 100 then x else x*2) + 1  
```

```haskell
doubleSmallNumber 39

doubleSmallNumber' 39

doubleSmallNumber 235.3

doubleSmallNumber' 235.3
```

```haskell
-- списки могут содердать только значения одного типа
[1,2,3,4]
:t [1,2,3,4]

[]
:t []

[1,2.5,3,4]
:t [1,2.5,3,4]

let b = [[1,2,3,4],[5,3,3,3],[1,2,2,3,4],[1,2,3]] 
:t b

['h','e','l','l','o']
:t ['h','e','l','l','o']
"hello" == ['h','e','l','l','o']
```

```haskell
-- добавление нового элемента в начало спика 
5:[1,2,3,4,5] 
:t (:)
```

```haskell
-- списки сравнимых элементов можно сравнивать, они сравниваютсяв лексикографическом порядке
[3,2,1] > [2,1,0]  
[3,2,1] > [2,10,100]  
[3,4,2] > [3,4] 
[3,4,2] > [2,4] 
[3,4,2] == [3,4,2]  
```

```haskell
-- так же для них есть множство полезных функций
head [5,4,3,2,1] 
:t head

tail [5,4,3,2,1] 
:t tail

last [5,4,3,2,1]
:t last

init [5,4,3,2,1] 
:t init

length [5,4,3,2,1]  
:t length

null [1,2,3] 
null [] 
:t null

elem 5 [1,2,3,4,5]
:t elem

reverse [5,4,3,2,1] 
:t reverse

take 3 [5,4,3,2,1]
take 1 [3,9,3] 
take 5 [1,2]
take 0 [6,6,6] 
:t take

drop 3 [8,4,2,1,5,6]  
drop 0 [1,2,3,4] 
drop 100 [1,2,3,4] 
:t drop

minimum [8,4,2,1,5,6] 
:t minimum

maximum [1,9,2,3,4] 
:t maximum

sum [5,2,1,6,3,2,5,7]  
:t sum

product [6,2,1,2] 
:t product
```

```haskell
--так же можно делать дипазоны
[1..20]  
['a'..'z']
['K'..'Z'] 
[0.1, 0.3 .. 1] 
-- и сложные диапазоны
[20,19..1]

--так же можно создавать бесконечные списки [1,2 ..]
take 10 [1,2..]
:t [1,2 ..]

take 10 (cycle [1,2,3])
take 12 (cycle "LOL ")  
:t cycle

take 10 (repeat 5) 
:t repeat
```
